```python
# 2. Find the most frequent words in a text file using sets
# 3. Filter user-submitted comments stored in a list for profanity
# 4. Build a histogram from word counts using dictionaries
# 5. Convert a deeply nested list to a flat list.


def most_frequent_words(file_path):
    with open(file_path, "r") as file:
        text = file.read().lower() # Read the file and convert to lowercase
    words = text.split()
    word_count = {}

    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    # Use a set to find unique words
    unique_words = set(words)
    most_frequent = sorted(unique_words, key=lambda x: word_count[x], reverse=True)

    return most_frequent[:20]  # Return the top 10 most frequent words

# Example usage:
file_path = '01-variables-and-assignments.ipynb'
print(most_frequent_words(file_path))
```

```python
#### Build a histogram from word counts using dictionaries. 

def build_histogram(text):
    # Split the text into words
    words = text.split()
    word_count = {}

    # Count the occurrences of each word
    for word in words:
        word = word.lower()  # Convert to lowercase for uniformity
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    return word_count

# Example usage:
text = "This is a sample text. This text is just a sample."
histogram = build_histogram(text)

# Display the histogram
for word, count in histogram.items():
    print(f'{word}: {count}')
```

```python
### Convert a deeply nested list to a flat list.


def flatten(nested_list):
    flat_list = []

    for item in nested_list:
        if isinstance(item, list):
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)
    return flat_list

nested_list = [1, [2,3], [4, [5,6]]]
print(flatten(nested_list))
# Output: [1, 2, 3, 4, 5, 6]
```
